##**VISUALIZACION DE LAS VENTAS POR CATEGORÍAS**

```python
import matplotlib.pyplot as plt
import pandas as pd

# --- Calcular la categoría más vendida por cada tienda ---
cat_mas1 = tienda['Categoría del Producto'].value_counts().idxmax()
cant_mas1 = tienda['Categoría del Producto'].value_counts().max()

cat_mas2 = tienda2['Categoría del Producto'].value_counts().idxmax()
cant_mas2 = tienda2['Categoría del Producto'].value_counts().max()

cat_mas3 = tienda3['Categoría del Producto'].value_counts().idxmax()
cant_mas3 = tienda3['Categoría del Producto'].value_counts().max()

cat_mas4 = tienda4['Categoría del Producto'].value_counts().idxmax()
cant_mas4 = tienda4['Categoría del Producto'].value_counts().max()

# --- Crear DataFrame resumen ---
data = {
    'Tienda': ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4'],
    'Categoría Más Vendida': [cat_mas1, cat_mas2, cat_mas3, cat_mas4],
    'Cantidad Vendida': [cant_mas1, cant_mas2, cant_mas3, cant_mas4]
}

df_popular = pd.DataFrame(data)

# --- Gráfico de barras horizontales ---
plt.figure(figsize=(8,6))
bars = plt.barh(df_popular['Tienda'], df_popular['Cantidad Vendida'], color='#27ae60', alpha=0.8)

# Agregar el valor dentro de la barra y la categoría afuera
for bar, (categoria, cantidad) in zip(bars, zip(df_popular['Categoría Más Vendida'], df_popular['Cantidad Vendida'])):
    # Posición del texto
    plt.text(bar.get_width() * 0.85, bar.get_y() + bar.get_height()/2,
             f"{cantidad:,}", va='center', ha='right', color='white', fontsize=10, fontweight='bold')

    # Categoría afuera de la barra
    plt.text(bar.get_width() + 10, bar.get_y() + bar.get_height()/2,
             f"{categoria}", va='center', fontsize=9, color='black')

# Título y etiquetas
plt.title('Categoría Más Popular por Tienda', fontsize=14)
plt.xlabel('Cantidad Vendida')
plt.ylabel('Tiendas')

plt.tight_layout()
plt.show()
```

##**VISUALIZACION DE LOS PRODUCTOS MÁS VENDIDOS Y MENOS VENDIDOS POR TIENDA**

```python
import matplotlib.pyplot as plt
import numpy as np

# Datos
mas_vendidos = [
    tienda.Producto.value_counts().idxmax(),
    tienda2.Producto.value_counts().idxmax(),
    tienda3.Producto.value_counts().idxmax(),
    tienda4.Producto.value_counts().idxmax()
]

menos_vendidos = [
    tienda.Producto.value_counts().idxmin(),
    tienda2.Producto.value_counts().idxmin(),
    tienda3.Producto.value_counts().idxmin(),
    tienda4.Producto.value_counts().idxmin()
]

cant_mas = [
    tienda.Producto.value_counts().max(),
    tienda2.Producto.value_counts().max(),
    tienda3.Producto.value_counts().max(),
    tienda4.Producto.value_counts().max()
]

cant_menos = [
    tienda.Producto.value_counts().min(),
    tienda2.Producto.value_counts().min(),
    tienda3.Producto.value_counts().min(),
    tienda4.Producto.value_counts().min()
]

tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Posiciones de las barras
y = np.arange(len(tiendas))
ancho = 0.35  # separación entre barras

plt.figure(figsize=(10,6))

# Gráfico de barras horizontales lado a lado
plt.barh(y - ancho/2, cant_mas, height=ancho, color='seagreen', alpha=0.8, label='Más vendido')
plt.barh(y + ancho/2, cant_menos, height=ancho, color='salmon', alpha=0.8, label='Menos vendido')

# Etiquetas con nombre y cantidad
for i in range(len(tiendas)):
    plt.text(cant_mas[i] + 1, y[i] - ancho/2, f"{mas_vendidos[i]} ({cant_mas[i]})", va='center', color='green', fontsize=9)
    plt.text(cant_menos[i] + 1, y[i] + ancho/2, f"{menos_vendidos[i]} ({cant_menos[i]})", va='center', color='darkred', fontsize=9)

# Configuración del gráfico
plt.yticks(y, tiendas)
plt.xlabel('Cantidad Vendida')
plt.ylabel('Tienda')
plt.title('Producto Más y Menos Vendido por Tienda')
plt.legend()
plt.tight_layout()
plt.show()

```
